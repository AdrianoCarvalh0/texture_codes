### 1 - Manual marking of vessels

Define in the <font color='skyblue'>main.py</font> file contained in the Annotation folder, the image you want to manually mark, the directory where the image is located and the output directory. The output directory will record the manual demarcations, creating the pair of coordinates resulting from the demarcation.

After running the code, scroll down to the last line of the file:

Dash app running on:

http://127.0.0.1:8050/


Click on the generated link. After that, the Dashboard will load.



Notice at the top center there is a toolbar:



Select the zoom tool. Use this tool, selecting the region in the image where the vessel to be demarcated is located.



In this example, I zoomed in until this vase was highlighted.



Note: if you want to zoom out again, just click on the "-" zoom button. As shown in the image below



After selecting the vessel to be demarcated, start the demarcation process:

Always start from left to right. Do the same for both sides of the vase.

Use the Draw Open FreeForm tool, as shown in the image below:



Pots must always be marked following the alignment of their edges. Neither leaving outside its limits nor entering its interior. The demarcation is for a part of the vessel. And the demarcation must not be made at the edge (corners) of the image or at the bifurcations (crossings) of the vessels.



After making all the demarcations for the image in question, following the recommendations mentioned above, the .json files will be saved in the desired output folder.

### 2 - Transformation of demarcations in Vessel Models

Configure the necessary parameters and then run the <font color='skyblue'><b>main.py</b></font> file contained in the <font color='orange'><b>Vessel_analysis</b> folder </font>. This file calls the function that reads the .json files, where the coordinates resulting from manual marking are located, and transforms them into map models.

### 3 - Insertion of created maps into backgrounds

Configure the existing parameters in the <font color='skyblue'><b>main.py</b></font> file contained in the <font color='orange'><b>Background</b></font>. This file calls the function that inserts the created vases, builds backgrounds, if desired, creates Bezier curves and records the backgrounds with vases as stipulated by the parameters.
