### Annotation

Creates an interface (Dashboard) where it is possible to demarcate one or more vessels. The demarcated vessels will be saved in a .json file that contains the coordinates of the demarcated polygons.




### Artificial_lines

<font color='orange'><b>Bezier_traces</b></font>: 500 traces in .json format whose Euclidean distance varies between 500 and 1300 pixels. These traces are used to transform artificial maps. The caliber of the lines depends on the number of lines on the map that will be "fitted"



### Background

<font color='orange'><b>Artificially_generated_maps</b></font>: There are 212 artificially generated backgrounds.



<font color='skyblue'><b>File background_generation.py</b></font> contains a set of functions that allow the insertion of maps in the background.

<font color='skyblue'><b>File main.py</b></font>  main element for inserting maps in the background. Modify the parameters as desired to create the artificial images.

### Trainning_validation

<font color="orange"><b>Artificial_Images</b></font> ==> Contains four sub-folders.

<font color="green"><b>Generated_from_1_map</b></font>. Contains 10 image packages, each containing 100 artificial images with their respective labels. The same map generated 100 images in 1 package. In total there were 10 different maps, where each one generated 100 images.

<font color="green"><b>Generated_from_5_maps</b></font> Contains 10 image packages, each containing 100 artificial images with their respective labels. 5 different maps generated 100 images in 1 package. In total there were 50 different maps, where each 5 generated 100 images.

<font color="green"><b>Generated_from_10_maps</b></font> Contains 1 image package, containing 100 artificial images with their respective labels. 100 images were generated using 10 randomly drawn maps.

<font color="green"><b>Generated_from_160_maps</b></font> Contains 1 image package, containing 100 artificial images with their respective labels. 100 images were generated using 160 randomly selected maps.

<font color="orange"><b>Backgrounds</b></font> ==> contains four sub-folders. One with 5, another with 10, another with 40 and another with 50 backgrounds.

<font color="orange"><b>Original_Images</b></font> ==> these are the 50 images used. This folder includes 40 training and 10 validation folders.

<font color="orange"><b>Labels_Original_Images</b></font> ==> there are 50 labels of original images.

<font color="orange"><b>Maps</b></font> ==> Contains four sub-folders.

<font color="green"><b>5_map_5_images</b></font>: there are 5 maps used to generate the 100 images in the Generated from 5 maps folder.

<font color="green"><b>10_maps_10_images</b></font> are 10 maps used to generate the 100 images in the Generated from 10 maps folder.

<font color="green"><b>160_maps_40_images</b></font> are 160 maps used to generate the 100 images in the Generated from 160 maps folder.

<font color="green"><b>200_maps_50_images</b></font> are the 4 best maps of the 50 images used for training and validation.

<font color="orange"><b>Maps_Plots</b></font> ==> Map plots. Only for viewing the extracted maps.

<font color="green"><b>Training </b></font> ==> Contains two sub-folders.
The first (images) contains 40 images for training. The second (labels) contains 40 labels of the respective training images.

<font color="green"><b>Validation</b></font> ==> Contains two subfolders.

The first (images) contains 10 images for validation.

The second (labels) contains 10 labels of the respective validation images.

### Utils

It has useful functions to make the code easier to use.
Being them:

<font color='skyblue'><b>checking_nans.py</b></font> Checks and displays null rows or columns

<font color='skyblue'><b>create_backgrounds.py</b></font> Creates and saves as many backgrounds as set

<font color='skyblue'><b>creating_bezier_curves.py</b></font> Creates and saves as many Bezier curves as set

<font color='skyblue'><b>display_artificial_maps.py</b></font> Display artificial backgrounds with vessel

<font color='skyblue'><b>display_backgrounds.py</b></font> Display artificial backgrounds

<font color='skyblue'><b>display_delauney.py</b></font> Displays and creates .svg images from Delauney triangulations

<font color='skyblue'><b>display_map_models_of_a_specific_image.py</b></font> Display map models of a specific image

<font color='skyblue'><b>display_traces.py</b></font> Reads the generated Bezier traces and creates plots of the traces and saves them later

<font color='skyblue'><b>display_vessel_maps_pickle.py</b></font> Display vessel_maps from pickle files

<font color='skyblue'><b>functions.py</b></font> Functions that read directories and create noise in images

<font color='skyblue'><b>generate_noise.py</b></font> Element that calls functions that generate noise and displays them

<font color='skyblue'><b>Image_properties.py</b></font> Display properties of images
